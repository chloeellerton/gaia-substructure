```python
### imports ###

from __future__ import division
from astropy.io import fits
import numpy
import matplotlib
from matplotlib import pylab, mlab, pyplot
from IPython.display import display
from IPython.core.pylabtools import figsize, getfigs
from pylab import *
from numpy import *
import scipy
import matplotlib.pyplot as pyplot
import scipy.integrate
from scipy.integrate import quad
from scipy.interpolate import UnivariateSpline
from matplotlib import gridspec
import scipy.optimize
from scipy.optimize import minimize
import scipy.stats
import matplotlib.ticker as ticker
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator)
from matplotlib import rc
from astropy.visualization import simple_norm
from astropy.table import Table
from scipy.optimize import curve_fit
from matplotlib.ticker import EngFormatter, StrMethodFormatter
import csv
from astropy import wcs
from astropy.io import fits
import datetime
import sys
import statistics
from scipy import ndimage, misc
import scipy.optimize as opt
from scipy.ndimage.measurements import label
from scipy.special import kv
from scipy.special import iv
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
import sqlutilpy
import healpy as hp
import pandas as pd
from astropy.coordinates import SkyCoord
from astropy.coordinates import ICRS, Galactic, FK4, FK5  
from astropy.coordinates import Angle, Latitude, Longitude 
import matplotlib.patches as patches
import sfdmap
import io, os, sys, types
from IPython import get_ipython
from nbformat import read
from IPython.core.interactiveshell import InteractiveShell
from csv import writer

### import module containing useful functions ###
# must be a way to tidy this up a bit

def find_notebook(fullname, path=None):
    """find a notebook, given its fully qualified name and an optional path

    This turns "foo.bar" into "foo/bar.ipynb"
    and tries turning "Foo_Bar" into "Foo Bar" if Foo_Bar
    does not exist.
    """
    name = fullname.rsplit('.', 1)[-1]
    if not path:
        path = ['']
    for d in path:
        nb_path = os.path.join(d, name + ".ipynb")
        if os.path.isfile(nb_path):
            return nb_path
        # let import Notebook_Name find "Notebook Name.ipynb"
        nb_path = nb_path.replace("_", " ")
        if os.path.isfile(nb_path):
            return nb_path
            
print(find_notebook('Cke_tools'))

class NotebookLoader(object):
    """Module Loader for Jupyter Notebooks"""
    def __init__(self, path=None):
        self.shell = InteractiveShell.instance()
        self.path = path

    def load_module(self, fullname):
        """import a notebook as a module"""
        path = find_notebook(fullname, self.path)

        print ("importing Jupyter notebook from %s" % path)

        # load the notebook object
        with io.open(path, 'r', encoding='utf-8') as f:
            nb = read(f, 4)


        # create the module and add it to sys.modules
        # if name in sys.modules:
        #    return sys.modules[name]
        mod = types.ModuleType(fullname)
        mod.__file__ = path
        mod.__loader__ = self
        mod.__dict__['get_ipython'] = get_ipython
        sys.modules[fullname] = mod

        # extra work to ensure that magics that would affect the user_ns
        # actually affect the notebook module's ns
        save_user_ns = self.shell.user_ns
        self.shell.user_ns = mod.__dict__

        try:
          for cell in nb.cells:
            if cell.cell_type == 'code':
                # transform the input to executable Python
                code = self.shell.input_transformer_manager.transform_cell(cell.source)
                # run the code in themodule
                exec(code, mod.__dict__)
        finally:
            self.shell.user_ns = save_user_ns
        return mod

class NotebookFinder(object):
    """Module finder that locates Jupyter Notebooks"""
    def __init__(self):
        self.loaders = {}

    def find_module(self, fullname, path=None):
        nb_path = find_notebook(fullname, path)
        if not nb_path:
            return

        key = path
        if path:
            # lists aren't hashable
            key = os.path.sep.join(path)

        if key not in self.loaders:
            self.loaders[key] = NotebookLoader(path)
        return self.loaders[key]

    
sys.meta_path.append(NotebookFinder())

import Cke_tools as cke 
```

```python
RRL_path = '/data/cke23/RRL_data/' # location of files 
```

```python
### read in original catalog with all RRL stars ###

all_stars = 'rrl_all_pm_error.fits'
hdul_data = fits.open(RRL_path+all_stars) 
data = hdul_data[1].data 
data_header = hdul_data[1].header 
data_cols = hdul_data[1].columns 
hdul_data.close()
```

```python
### find stars in distance range and make a new catalog ###

dist_all = data.field('DIST') 
print('number of stars in original file = {}'.format(len(dist_all)))

dist_min = 30
dist_indices = [i for i, x in enumerate(dist_all) if dist_min <= x ] 
print('number of stars in this with dist > {}'.format(dist_min) + ' kpc = {}'.format(len(dist_indices)))

table = Table.read(RRL_path+all_stars, hdu=1) 
table_dist_cut = table[dist_indices] 

dist_cut_stars = '>'+str(dist_min)+'kpc_RRL_dist.fits'
table_dist_cut.write(RRL_path+dist_cut_stars, overwrite=True) 
```

```python
### open fits catalog only with stars in distance range ###

hdul_dist_cut = fits.open(RRL_path+dist_cut_stars) 
data_dist_cut = hdul_dist_cut[1].data 
data_header_dist_cut = hdul_dist_cut[1].header
data_cols_dist_cut = hdul_dist_cut[1].columns 
hdul_dist_cut.close()

### data from stars in distance range ###

dist_kpc = data_dist_cut.field('DIST') #kpc
ra = data_dist_cut.field('RA') 
dec = data_dist_cut.field('DEC')
pmra = data_dist_cut.field('PMRA') 
pmdec = data_dist_cut.field('PMDEC') 
pmra_err = data_dist_cut.field('PMRA_ERROR') 
pmdec_err = data_dist_cut.field('PMDEC_ERROR') 
```

```python
### open csv catalog containing location of all known satellites, add extra column of mask sizes for each sat ###

sat_path = '/data/cke23/satellite_data/'
satellites_gabriel = 'CompiledSatCatalogv2_gabriel.csv'
colnames = ['name', 'T', 'ra', 'dec', 'Dh', 'Dhe', 'rh', 'rhe', 'MV', 'MVe', 'RV_h', 'RVe', 'ref']
data_sat = pd.read_csv(sat_path+satellites_gabriel, names=colnames, header=0)
names = data_sat.name.tolist() 

LMC_loc, SMC_loc, SAG_loc = names.index('LMC'), names.index('SMC'), names.index('SagittariusdSph')
data_sat['mask_size'] = 1.5 # most satellites need a mask radius of 1.5 deg
data_sat.loc[[LMC_loc, SMC_loc, SAG_loc], 'mask_size'] = 25, 15, 15 # some satellites need larger mask radius

satellites_all = 'sat_catalog_chloe.csv'
data_sat.to_csv(sat_path+satellites_all)

#need to add Antlia 2 to known satellites list
ant2_row = [300, 'Antlia2', 'g', 143.8868, -36.7673, '-', '-', '-', '-', '-', '-', '-', '-', 'Torrealba2016', 1.5]
with open(sat_path+satellites_all, 'a+', newline='') as write_obj:
    csv_writer = writer(write_obj)
    csv_writer.writerow(ant2_row)
    
colnames_mask = ['name', 'T', 'ra', 'dec', 'Dh', 'Dhe', 'rh', 'rhe', 'MV', 'MVe', 'RV_h', 'RVe', 'ref', 'mask_size']
sat_cat_masks = pd.read_csv(sat_path+satellites_all, names=colnames_mask, header=0) 

mask_size = sat_cat_masks.mask_size.tolist()
sat_ra = sat_cat_masks.ra.tolist()
sat_dec = sat_cat_masks.dec.tolist()
print(sat_cat_masks)
```

```python
### take ra, dec and mask sizes, produce a map where all known satellites are masked ###

healpix_path = '/data/cke23/healpix_maps/'

nside = 64
npix = hp.nside2npix(nside) 
vec_sat = cke.radec2vec(sat_ra, sat_dec)
map_sat_loc = hp.ma(np.arange(hp.nside2npix(nside), dtype=np.double))
mask_sat_loc = np.zeros(hp.nside2npix(nside), dtype=np.bool)

for i in range(len(mask_size)):
    masked = hp.query_disc(nside, vec_sat[i], np.deg2rad(mask_size[i]), inclusive=True)
    mask_sat_loc[masked] = 1
map_sat_loc.mask = mask_sat_loc
hp.mollview(map_sat_loc)
sat_masks_hp_map = 'satellite_masks_map.fits'
hp.write_map(healpix_path+sat_masks_hp_map, map_sat_loc, overwrite=True)
```

```python
### make healpix map of all stars in the RRL catalog greater than min distance ###

theta, phi = cke.radec2thetaphi(ra,dec)
indices = hp.ang2pix(nside, theta, phi)
idx, _, _, counts = np.unique(indices, return_index=True, return_inverse=True, return_counts=True)
RRL_map = np.zeros(npix, dtype=int)
RRL_map[idx] = counts

hp.mollview(np.log10(RRL_map+1)) 
RRL_hp_map = '>'+str(dist_min)+'kpc_RRL_map.fits'
hp.write_map(healpix_path+RRL_hp_map, RRL_map, overwrite=True) 

table_indexed = Table.read(RRL_path+dist_cut_stars, hdu=1) 
table_indexed['PIXEL']=indices
dist_cut_stars_indexed = 'nside_'+str(nside)+'_>'+str(dist_min)+'kpc_RRL_dist.fits'
table_indexed.write(RRL_path+dist_cut_stars_indexed, overwrite=True) 
```

```python
### apply the mask ###

RRL_map_masked = RRL_map*1
for i in range(len(mask_size)):
    masked = hp.query_disc(nside, vec_sat[i], np.deg2rad(mask_size[i]), inclusive=True)
    RRL_map_masked[masked] = 0   
hp.mollview(np.log10(RRL_map_masked+1)) 
RRL_hp_map_masked = '>'+str(dist_min)+'kpc_RRL_map_satellites_masked.fits'
hp.write_map(healpix_path+RRL_hp_map_masked, RRL_map_masked, overwrite=True)
```

```python
### now mask over the Sag stream ###

# range of sag coordinates to mask over
lam_min, lam_max = -np.deg2rad(180), +np.deg2rad(180)
b_min, b_max = -np.deg2rad(15), +np.deg2rad(15)
lam_step, b_step = np.deg2rad(0.5), np.deg2rad(0.5)

lam,b = np.mgrid[lam_min:(lam_max+lam_step):lam_step, b_min:(b_max+b_step):b_step]
lam,b = lam.flatten(), b.flatten()
ra_sag, dec_sag = cke.sag_coord2radec(lam, b)[0], cke.sag_coord2radec(lam, b)[1]

RRL_map_masked_sag = RRL_map_masked*1 # make a copy of the previous map 
RRL_map_masked_sag = cke.hpx_mask(RRL_map_masked_sag, np.rad2deg(ra_sag), np.rad2deg(dec_sag))

hp.mollview(np.log10(RRL_map_masked_sag+1))
RRL_hp_map_masked_sag = '>'+str(dist_min)+'kpc_RRL_map_satellites_sag_masked.fits'
hp.write_map(healpix_path+RRL_hp_map_masked_sag, RRL_map_masked_sag, overwrite=True)
```
